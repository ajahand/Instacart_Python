```python
# import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import scipy
```

## 3.create a histogram of order time

```python
#creating a histogram

plt.figure(figsize=(8, 6))  

# Plotting histogram with customized parameters
hist = df_instacart['order_time'].plot.hist(bins=24, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Order Time')
plt.ylabel('Frequency')
plt.title('Distribution of Order Time')

plt.grid(axis='y') 
plt.show()
```

Most order occur between 9a thru 3pm

Not a lot of orders between 12 am and 6am 

order frequency decreases gradually decreases after peak hours


## 4. create a bar chart for loyalty flags



```python
#creating a bar chart

plt.figure(figsize=(8, 6))  # Adjust figure size if needed

# Plotting bar chart with customized parameters
loyalty_flag = df_instacart['loyalty_flag'].value_counts().plot(kind='bar', color='green')

# Adding labels and title
plt.xlabel('Loyalty Flag')
plt.ylabel('Count')
plt.title('Count of Loyalty Flags')

plt.xticks(rotation=0)  # Adjust rotation of x-axis labels if necessary
plt.grid(axis='y')  # Adding grid lines for better visualization
plt.show()
```


## 5. create a line chart and sample data


```python
#creating sample data
np.random.seed(4)
dev = np.random.rand(len(df_instacart)) <= 0.7
```

```python
#assigning parts of the numbers that are <=0.7 to small and numbers >=0.7 to big
big = df_instacart[dev]
small = df_instacart[~dev]
```

```python
#checking the len
len(df_instacart)
```

```python
#making sure the small and big add to the whole df
len(big)+len(small)
```

```python
#creating a df that just grabs whats needed for line chart
price_time = df_2 = small[['mean_spent','order_time']]
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed

# Plotting line chart with customized parameters
line = sns.lineplot(data=df_2, x='order_time', y='mean_spent', color='orange', marker='o')

# Adding labels and title
plt.xlabel('Order Time')
plt.ylabel('Revenue')
plt.title('Revenue Over Time')

plt.grid(True)  # Adding grid lines for better visualization
plt.show()
```


## 7. create a line chart for the age to family situation


```python
#creating a df that just grabs whats needed for line chart
df_3 = small[['age','n_dependents']]
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed

# Creating a scatterplot instead of a line plot with customized parameters
line_2 = sns.scatterplot(data=df_3, x='age', y='n_dependents', color='green', marker='o')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Number of Dependents')
plt.title('Relation between Age and Number of Dependents')

plt.grid(True)  # Adding grid lines for better visualization
plt.show()
```

```python
#creating a df that just grabs whats needed for scatterplot
df_4 = small[['age','income']]
```


## 8. create a scatterplot for the correlation between age and income

```python
scatter =sns.scatterplot(x = 'age', y = 'income',data = df_4)
```

No correlation found

```python
# Filter data for age 0 to 40 and age 41 and up
df_4_age_0_to_40 = df_4[df_4['age'] <= 40]
df_4_age_41_and_up = df_4[df_4['age'] > 40]

plt.figure(figsize=(12, 6))  # Adjust figure size if needed

# Creating scatter plot for age 0 to 40
plt.subplot(1, 2, 1)  # Creating the first subplot
scatter_age_0_to_40 = sns.scatterplot(x='age', y='income', data=df_4_age_0_to_40, color='blue')
plt.title('Age 0-40 vs Income')
plt.xlabel('Age')
plt.ylabel('Income')

# Creating scatter plot for age 41 and up
plt.subplot(1, 2, 2)  # Creating the second subplot
scatter_age_41_and_up = sns.scatterplot(x='age', y='income', data=df_4_age_41_and_up, color='red')
plt.title('Age 41+ vs Income')
plt.xlabel('Age')
plt.ylabel('Income')

plt.tight_layout()  # Adjust subplots to prevent overlap
plt.show()
```
