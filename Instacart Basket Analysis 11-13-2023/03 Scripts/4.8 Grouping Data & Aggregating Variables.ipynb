```python
#4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# 5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). 
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
#Create a spending flag for each user based on the average price across all their orders using the following criteria:
df_ords_prods_merge['mean_spent'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_spent'] >= 10, 'spending_flag'] = 'High Spender'
df_ords_prods_merge.loc[df_ords_prods_merge['mean_spent'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#7. Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column.
df.groupby('days_since_prior_order').agg({'order_number': ['median']})
```

```python
df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] <= 20) & (df_ords_prods_merge['order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
